```python
# importing the required packages

import re
import pandas as pd 
import mysql.connector as db
```

```python
# reading the csv files

customers = pd.read_csv('Customers.csv', encoding='ISO-8859-1')
exchange_rates = pd.read_csv('Exchange_Rates.csv', encoding='ISO-8859-1')
products = pd.read_csv('Products.csv', encoding='ISO-8859-1')
sales = pd.read_csv('Sales.csv', encoding='ISO-8859-1')
stores = pd.read_csv('Stores.csv', encoding='ISO-8859-1')
```

## Cleaning the tables ## 

```python
customers.info()
```

```python
# converting the datatype of birthday 
customers['Birthday'] = pd.to_datetime(customers['Birthday'])
```

```python
customers['CustomerKey'] = customers['CustomerKey'].fillna(0)
customers['Gender'] = customers['Gender'].fillna(0)
customers['Name'] = customers['Name'].fillna(0)
customers['City'] = customers['City'].fillna(0)
customers['State Code'] = customers['State Code'].fillna(0)
customers['State'] = customers['State'].fillna(0)
customers['Zip Code'] = customers['Zip Code'].fillna(0)
customers['Country'] = customers['Country'].fillna(0)
customers['Continent'] = customers['Continent'].fillna(0)
customers['Birthday'] = customers['Birthday'].fillna(0)
```

```python
customers.info()
```

```python
# converting the dtype of date in exchange_rates
exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'])
```

```python
exchange_rates['Date'] = exchange_rates['Date'].fillna(0)
exchange_rates['Currency'] = exchange_rates['Currency'].fillna(0)
exchange_rates['Exchange'] = exchange_rates['Exchange'].fillna(0)
```

```python
exchange_rates.info()
```

```python
products.info()
```

```python
products['ProductKey'] = products['ProductKey'].fillna(0)
products['Product Name'] = products['Product Name'].fillna(0)
products['Brand'] = products['Brand'].fillna(0)
products['Color'] = products['Color'].fillna(0)
products['Unit Cost USD'] = products['Unit Cost USD'].fillna(0)
products['Unit Price USD'] = products['Unit Price USD'].fillna(0)
products['SubcategoryKey'] = products['SubcategoryKey'].fillna(0)
products['Subcategory'] = products['Subcategory'].fillna(0)
products['CategoryKey'] = products['CategoryKey'].fillna(0)
products['Category'] = products['Category'].fillna(0)
```

```python
def extract_value(x):
    pattern = r'[0-9]+\.[0-9]{0,4}'
    result = re.search(pattern, x)
    return result.group()
```

```python
products['Unit Cost USD'] = products['Unit Cost USD'].apply(extract_value)
```

```python
products['Unit Cost USD'] = products['Unit Cost USD'].astype(float)
```

```python
# converting unit price usd to int

products['Unit Price USD'] = products['Unit Price USD'].apply(extract_value)
products['Unit Price USD'] = products['Unit Price USD'].astype(float)
```

```python
products.info()
```

```python
# cleaning datas from sales dataframe 

sales.info()
```

```python
sales['Order Number'] = sales['Order Number'].fillna(0)
sales['Line Item'] = sales['Line Item'].fillna(0)
sales['Order Date'] = sales['Order Date'].fillna(0)
sales['Delivery Date'] = sales['Delivery Date'].fillna(0)
sales['CustomerKey'] = sales['CustomerKey'].fillna(0)
sales['StoreKey'] = sales['StoreKey'].fillna(0)
sales['ProductKey'] = sales['ProductKey'].fillna(0)
sales['Quantity'] = sales['Quantity'].fillna(0)
sales['Currency Code'] = sales['Currency Code'].fillna(0)
```

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
```

```python
sales['Delivery Date'] = pd.to_datetime(sales['Delivery Date'])
```

```python
sales.info()
```
